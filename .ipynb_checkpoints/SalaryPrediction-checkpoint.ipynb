```python
#this class can be used the load the data into a dataframe and also gives the shape of the dataframe
class LoadManager():
    def __init__(self, train_features_path, test_features_path, train_labels_path):
        self.train_features_path = train_features_path
        self.test_features_path = test_features_path
        self.train_labels_path = train_labels_path
    def fit(self):
        return self
    def load_data(self, path):
        print("Loading sample data from", path, "into a dataframe")
        dataFrame = pd.read_csv(path)
        print("done..")
        print("here is the sample data from the dataFrame \n")
        print(dataFrame.head())
        print(" \n shape of the data frame is: ", dataFrame.shape)
        return dataFrame
    def transform(self):
        train_features_df = self.load_data(self.train_features_path)
        test_features_df = self.load_data(self.test_features_path)
        train_labels_df = self.load_data(self.train_labels_path)
        return train_features_df, test_features_df, train_labels_df
   
```

```python
#This class merges the data based on the unique identifier, 
#binns the numerical variable into a categorical variable and 
#creates a correlation heatmap using heatmap
class BasicDataProcessing():
    def __init__(self, train, labels = None):
        self.train = train
        self.labels = labels        
    def merge_data(self, unique_identifier): #merging the data based on the unique identifier
        merge_df = pd.merge(self.train, self.labels, on = unique_identifier)
        print("merging done")
        print("here is the sample data from the new dataFrame \n")
        print(merge_df.head())
        print(" \n shape of the data frame is: ", merge_df.shape)
        return merge_df
    def binning(self, target ,num_of_samples, group_names): #binning the numerical variable into a categorical variable for easier analysis
        print("Performing binning")
        bins = np.linspace(min(self.train[target]), max(self.train[target]), num_of_samples)
        target_binned = target + "Binned"
        print("Adding the binned column to the dataFrame")
        self.train[target_binned] = pd.cut(self.train[target], bins, labels = group_names, include_lowest = True)
        return self.train
    def correlation_heatMap(self): #creating a correlation heatmap using heatmap
        corr = self.train.corr()
        sns.heatmap(corr, 
                xticklabels=corr.columns.values,
                yticklabels=corr.columns.values)
        plt.title('Heatmap of Correlation Matrix')
        corr
        plt.gcf().set_size_inches(11.7, 8.27)
```

```python
#This class helps to find duplicate data, missing data in the train or test data and 
#detecting outliers in the target variable using boxplot.

class FindingInaccurateData():
    def __init__(self, train, test):
        self.train = train
        self.test = test
 
    def fit(self):
        return self
    def transform(self, target):
        print("***** Looking for duplicate data ******")
        print("looking for duplicate data in train data \n")
        train_duplicate_data_sum = self.train.duplicated().sum()
        print("There are", train_duplicate_data_sum, "duplicate records in the train data")
        print("looking for duplicate data in test data \n")
        test_duplicate_data_sum = self.test.duplicated().sum()
        print("There are",  test_duplicate_data_sum, "duplicate records in the test data")
        
        print("\n **** Detecting Missing Data *******")
        print("Checking the missing data in the train data \n")
        train_missing_Data_Sum = self.train.isnull().sum().sum()
        print("There are",  train_missing_Data_Sum, "missing records in the test data")
        print("Checking the missing data in the test data \n")
        test_missing_Data_Sum = self.test.isnull().sum().sum()
        print("There are",  test_missing_Data_Sum, "missing records in the test data")
        
        print("\n ***** Detecting Outliers in the target of train dataframe *******")
        print("Plotting the box plot of given target to detect outliers")
        sns.boxplot(x = self.train[target])
        
        
    
  
```

```python

#This class gives the basic statistics of numerical and categorical columns present in the data frame 
#given as an input and it also summarizes the target varaible.
BasicStatistics():
    def __init__(self, df, target = None):
        self.df = df
        self.target = target
    def fit(self):
        return self
    def transform(self, include_target = False):
        print("Numerical cols statistics \n")
        print(self.df.describe(include = [np.number]))
        print("\n Categorical cols statistics \n")
        print(self.df.describe(include = ['O']))
        if include_target == True:
            print("\n summarizing the target variable")
            group_By_Target = self.df.groupby(self.target)
            print(group_By_Target.mean())
```

```python

#This class can be used to create a baseline model and 
#plot validation curve between training error and vallidation error of the baseline model. 

class BaselineModel():
    def __init__(self,train, labels, baseLineModel = LinearRegression(), 
                 noOfFolds = 5, performanceMeasure = 'neg_mean_squared_error'):
        self.train = train
        self.labels = labels
        self.baseLineModel = baseLineModel
        self.noOfFolds = noOfFolds
        self.performanceMeasure = performanceMeasure
    
    def fit(self):
        return self
    
    def plot_Val_Curve(self,train_sizes):
        train_sizes, train_scores, validation_scores = learning_curve(
                                 estimator = self.baseLineModel,
                                      X = self.train,
         y = self.labels, train_sizes = train_sizes,
                scoring = self.performanceMeasure, cv= self.noOfFolds)
        print('Training scores:\n\n', train_scores)
        print('\n', '-' * 70) # separator to make the output easy to read
        print('\nValidation scores:\n\n', validation_scores)
        print("\n Taking the mean value of each row and also flip the signs of the error scores. \n")
        train_scores_mean = -train_scores.mean(axis = 1)
        validation_scores_mean = -validation_scores.mean(axis = 1)

        print('Mean training scores\n\n', pd.Series(train_scores_mean, index = train_sizes))
        print('\n', '-' * 20) # separator
        print('\nMean validation scores\n\n',pd.Series(validation_scores_mean, index = train_sizes))
        
        print("\n Plotting the Validation Curve for Baseline Model")
        plt.style.use('seaborn')

        plt.plot(train_sizes, train_scores_mean, label = 'Training error')
        plt.plot(train_sizes, validation_scores_mean, label = 'Validation error')

        plt.ylabel('MSE', fontsize = 14)
        plt.xlabel('Training set size', fontsize = 14)
        plt.title('Learning curves for a linear regression model', fontsize = 18, y = 1.03)
        plt.legend()
            
    
    def transform(self, train_sizes):
        print("Perform cross validation on baseline model with ", self.noOfFolds, "folds")
        baseLineModelScores = cross_val_score(self.baseLineModel, 
                                              self.train, 
                                              self.labels, scoring = self.performanceMeasure , cv= self.noOfFolds)
        print("The baseLineModelScores during cross validation are \n")
        print(baseLineModelScores)
        print("\n")
        
        self.plot_Val_Curve(train_sizes)
        
        
```

```python
#This class creates a pipeline to process the data and converts the processed data into a n-dimensional array.
class DataManager(BaseEstimator, TransformerMixin): 
    def __init__(self, pipeline):
        self.pipeline = pipeline
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        data_prepared = self.pipeline.fit_transform(X)
        return data_prepared.toarray()
```

```python
#the model manager manages a list of models, cycles through them and keeps the best performing model
class ModelManager():
    def __init__(self, models,  scoring ='neg_mean_squared_error', kfold =5):
        self.models = models
        self.scoring = scoring
        self.kfold = kfold
    def fit(self, X, y):
        return self
    def transform(self, X, y):
        results = []
        names = []
        model_scoring_mean = {}
        for name, model in self.models:
            print('Managing', name)
            cv_results = model_selection.cross_val_score(model, X, y, cv=self.kfold, scoring=self.scoring)
            print('Cross Validation done, storing the results of', name)
            results.append(cv_results)
            names.append(name)
            print('Calculating and storing the mean of cv results for', name)
            model_scoring_mean[name] = cv_results.mean()
            msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
            print(msg)
        print('selecting the best model')
        best_model = max(model_scoring_mean, key=model_scoring_mean.get)  
        print(best_model, model_scoring_mean[best_model])
```
